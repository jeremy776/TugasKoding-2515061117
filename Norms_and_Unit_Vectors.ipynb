```python
import numpy as np
```

```python
# buat vector x dengan ukuran 3x1
x = np.array([[117], [171], [711]])
x
```

```python
# hitung nilai L2 norm dari vector x
x_l2 = np.linalg.norm(x)
x_l2
```

```python
# buat vector v dengan elemen (3,4)
v = np.array([[24, 30, 60, 10],
              [55, 34, 67, 77],
              [12, 34, 56, 78]])
v.shape
```

```python
# hitung nilai L2 norm dari vector v
v_l2 = np.linalg.norm(v)
v_l2
```

```python
# tampilkan unit vector dari vector v
v_unit = v_l2 / v
v_unit
```

```python
# hitung panjang unit vector dari vector v
v_unit_l2 = np.linalg.norm(v_unit)
v_unit_l2
```
