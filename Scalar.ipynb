```python
# Soal Tensorflow Scalar
import tensorflow as tf
```

```python
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow
x_tf = tf.Variable(117, dtype=tf.int16)
x_tf
```

```python
y_tf = tf.Variable(171, dtype=tf.int16)
y_tf
```

```python
tf.add(x_tf, y_tf)
```

```python
x_tf_float = tf.Variable(11.7)
x_tf_float
```

```python
tf.add(x_tf_float, y)
```
