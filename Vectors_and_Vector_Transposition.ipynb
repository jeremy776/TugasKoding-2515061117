```python
import numpy as np
```

```python
# buat vector x dengan ukuran 3x1
x = np.array([[117], [171], [711]])
x
```

```python
# hitung panjang vector x
panjang = np.linalg.norm(x)
panjang
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
x_t
```

```python
# tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[117], [171], [711]])
y_t = y.T
y_t
```

```python
# lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import torch
import tensorflow as tf
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t (Pytorch)
x_pt = torch.tensor([[117], [171], [711]])
x_pt_t = x_pt.T
x_pt_t
```

```python
# tampilkan ukuran dari vector x_t (PyTorch)
x_pt_t.shape
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t (Tensorflow)
x_tf = tf.Variable([[117], [171], [711]])
tf.transpose(x_tf)
```

```python
# tampilkan ukuran dari vector x_t (Tensorflow)
tf.shape(x_tf)
```
